# Read Excel File

```python
import pandas as pd

df = pd.read_excel('Scopus2018.xlsx', sheet_name='2018 All', header=1)
df.head()
```

# Count Data

```python
df.count()
```

# Count Journal

```python
allJournal = df.loc[df['Type'] == "Journal"]
allJournal.count()
```

```python
World_unique = allJournal.groupby('Scopus Source ID').sum()
World_unique.count()
```

# Malaysia

```python
pd.set_option('display.max_rows', 200)
Malaysia = allJournal[allJournal['Publisher'].str.contains('Malay')]
Malaysia = Malaysia[['Scopus Source ID','Title','Publisher','Quartile','Scopus Sub-Subject Area']]
Malaysia.sort_values('Quartile', ascending=True)
```

```python
Malaysia.count()
```

```python
Malaysia_unique = Malaysia.groupby(['Scopus Source ID']).sum()
Malaysia_unique.count()
```

```python
Malaysia_unique = Malaysia.groupby(['Scopus Source ID','Title','Publisher','Quartile']).sum()
Malaysia_unique.sort_values('Quartile', ascending=True)
```

```python
Malaysia_Q1 = Malaysia.loc[Malaysia['Quartile'] == "Quartile 1"]
Malaysia_Q1
```

```python
Total_Q1 = Malaysia_Q1.groupby(['Scopus Source ID']).sum()
Total_Q1.count()
```

```python
Malaysia_Q2 = Malaysia.loc[Malaysia['Quartile'] == "Quartile 2"]
Total_Q2 = Malaysia_Q2.groupby(['Scopus Source ID']).sum()
Total_Q2.count()
```

# UNIVERSITI UTARA MALAYSIA

```python
UUM = Malaysia[Malaysia['Publisher'].str.contains('Utara')]
UUM
```

```python
UUM_unique = UUM.groupby(['Scopus Source ID','Title','Publisher','Quartile']).sum()
UUM_unique.sort_values('Quartile', ascending=True)
```
